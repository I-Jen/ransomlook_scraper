```python
def is_utc_after_2023_epoch(utc_value_epoch):
  """Returns True if the UTC value in epoch time is before 2023, False otherwise."""

  utc_datetime = datetime.fromtimestamp(utc_value_epoch / 1000.0)
  jan_1_2023_datetime = datetime(2023, 1, 1)

  return utc_datetime >= jan_1_2023_datetime
```
